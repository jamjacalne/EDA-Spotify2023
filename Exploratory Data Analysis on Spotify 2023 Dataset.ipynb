# **Temporal Trends**


### Number of tracks released per year

```python
tracks_per_year = dataFrame['released_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(tracks_per_year.index, tracks_per_year.values)
plt.title('Number of Tracks Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.grid(True)
plt.show()
```

### Number of tracks released per month

```python
tracks_per_month = dataFrame['released_month'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(tracks_per_month.index, tracks_per_month.values)
plt.title('Number of Tracks Released Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Tracks')
plt.xticks(range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

# **Genre and Music Characteristics**

### Correlation between streams and musical attributes

```python
attributes = dataFrame[['streams', 'bpm', 'danceability_%', 'energy_%', 'valence_%', 'acousticness_%']]
correlation_matrix = attributes.corr()
print(correlation_matrix)

plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix (Streams v. Musical Attributes)")
plt.show()

streams_corr = correlation_matrix['streams'].drop('streams').sort_values(ascending=False)
print("Attributes most correlated with streams:")
print(streams_corr)

danceability_energy = correlation_matrix.loc['danceability_%', 'energy_%']
valence_acousticness = correlation_matrix.loc['valence_%', 'acousticness_%']
print(f"\nCorrelation between danceability_% and energy_%: {danceability_energy}")
print(f"Correlation between valence_% and acousticness_%: {valence_acousticness}")
```

# **Platform Popularity**

### Comparison of tracks across different platforms

```python
platform_sums = dataFrame[['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists']].sum()

plt.figure(figsize=(12, 6))
platform_sums.plot(kind='bar')
plt.title("Total Track Appearances in Spotify Playlists, Spotify Charts, and Apple Playlists")
plt.xlabel("Platform")
plt.ylabel("Total Appearances")
plt.xticks(rotation=0)
plt.show()

print("Total number of tracks in each platform:")
for platform, total in platform_sums.items():
    print(f"{platform.replace('_', ' ').title()}: {total}")
```

# **Advanced Analysis**

### Patterns among tracks with the same key or mode

```python
streams_key = dataFrame.groupby('key')['streams'].mean().sort_values(ascending=False)
print("Average streams per key:")
print(streams_key)

streams_mode = dataFrame.groupby('mode')['streams'].mean()
print("\nAverage streams per mode (1 = Major, 0 = Minor):")
print(streams_mode)
```

### Most frequently appearing artists in playlists or charts

```python
chart_counts = dataFrame.groupby('artist(s)_name')[['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists']].sum()

chart_counts['total_appearances'] = chart_counts.sum(axis=1)

top_artists_in_playlists_charts = chart_counts.sort_values(by='total_appearances', ascending=False).head()
print(top_artists_in_playlists_charts)
```
